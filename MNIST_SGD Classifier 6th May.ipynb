```python
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\icath\OneDrive\Documents\Salty Coder\mnist.csv')
df
```

```python
X = df.drop(['id','class'], axis=1).values # get just pixels
y = df['class'].values              # get just digit value
```

```python
X.shape
```

```python
#Split training and test data
X_train, X_test, y_train, y_test = X[:1600], X[1600:], y[:1600], y[1600:]
```

```python
sgd_classifier = SGDClassifier(random_state=42)
sgd_classifier.fit(X_train, y_train)
```

```python
y_predict=sgd_classifier.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_predict)
accuracy
```

```python
def display_images(images, predictions, n_rows=5, n_cols=5):
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 10))
    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i].reshape(28, 28), cmap='binary')
        ax.set_title(f'Prediction: {predictions[i]}', fontsize=10)
        ax.axis('off')
    plt.show()
```

```python
display_images(images=X_test, predictions=y_predict)
```
